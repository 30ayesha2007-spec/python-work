```python
#Answer question 1
marks = [85, 78, 92, 88, 80]

print("Marks of 5 subjects:")
for m in marks:
    print(m)

total = sum(marks)
print("Total Marks:", total)

average = total / len(marks)
print("Average Marks:", average)
```

```python
#Answer question 4
marks = {
    "arisha": [76, 78, 89, 77, 56],
    "samad": [68, 78, 77, 88 ,75],
    "basit": [89, 67, 73, 71, 85]
}
for name, marks in marks.items():
    total_marks = sum(marks)
    percentage = (total_marks / (len(marks) * 100)) * 100
    print(f"{name} Percentage: {percentage:.2f}%")
```
